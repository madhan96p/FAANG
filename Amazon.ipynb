# Standarding

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

Columns = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']
dd[Columns] = scaler.fit_transform(dd[Columns])
```

## Date/Time Features

```python
dd['Date'] = pd.to_datetime(dd['Date'])

dd['Year'] = dd['Date'].dt.year
dd['Month'] = dd['Date'].dt.month
dd['Day'] = dd['Date'].dt.day
dd['Weekday'] = dd['Date'].dt.weekday  # 0 = Monday, 6 = Sunday
dd['Is_Weekend'] = dd['Weekday'].isin([5, 6])  # True if Saturday or Sunday
# Just checking for is there any wrong data
```

```python
dd['Is_Weekend'].value_counts()
```

```python
dd.drop(columns=['Is_Weekend'], inplace=True)
```

## Return Features
*predicting future stock prices.*

```python
dd['Prev_Close'] = dd['Close'].shift(1)
dd.loc[dd['Date'] == '2005-01-03', 'Prev_Close'] = -0.857272 # First day has no previous close
dd['Return'] = (dd['Close'] - dd['Prev_Close']) / dd['Prev_Close'] * 100
```

# Dedecting outliers
**Purpose:** Outliers can affect the performance of machine learning models, especially regression models.

```python
# Dedecting outliers using Z-score
import numpy as np
from scipy import stats

# Z-score method to detect outliers
z_scores = stats.zscore(dd[Columns])
abs_z_scores = np.abs(z_scores)
dd_no_outliers = dd[(abs_z_scores < 3).all(axis=1)]  # Keep rows where Z-score < 3
dd_outliers = dd[(abs_z_scores >= 3).any(axis=1)]  # Keep rows where Z-score >= 3
```

# Exploratory Data Analysis (EDA)

## Data Visualization

### Matplotlib

#### Line Chart

```python
import matplotlib.pyplot as plt

dd['Date'] = pd.to_datetime(dd['Date'])
plt.figure(figsize=(12, 6))

plt.plot(dd['Date'], dd['Open'], label='Open Price', color='blue')
plt.plot(dd['Date'], dd['Close'], label='Close Price', color='green')

plt.title('Stock Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

#### Correlation Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = dd[['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']].corr()

plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title('Correlation Heatmap of Stock Features')
plt.tight_layout()
plt.show()
```

#### Box Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))

plt.subplot(1, 2, 1)
sns.boxplot(x=dd['Close'], color='green')
plt.title('Box Plot of Close Prices')

plt.subplot(1, 2, 2)
sns.boxplot(x=dd['Volume'], color='orange')
plt.title('Box Plot of Volume')

plt.tight_layout()
plt.show()
```

#### Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
sns.scatterplot(x=dd['Volume'], y=dd['Close'], color='purple')
plt.title('Relationship Between Volume and Close Price')
plt.xlabel('Volume')
plt.ylabel('Close Price')
plt.tight_layout()
plt.show()
```
