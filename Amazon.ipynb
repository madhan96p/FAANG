```python
dd['Prev_Close'] = dd['Close'].shift(1)
dd.loc[dd['Date'] == '2005-01-03', 'Prev_Close'] = -0.857272 # First day has no previous close
dd['Return'] = (dd['Close'] - dd['Prev_Close']) / dd['Prev_Close'] * 100
```

# Dedecting outliers
**Purpose:** Outliers can affect the performance of machine learning models, especially regression models.

```python
import numpy as np
import pandas as pd
from scipy.stats import zscore

columns_to_check = ['Open', 'High', 'Low', 'Close', 'Volume']

Q1 = dd[columns_to_check].quantile(0.25)
Q3 = dd[columns_to_check].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_iqr = (dd[columns_to_check] < lower_bound) | (dd[columns_to_check] > upper_bound)

z_scores = np.abs(zscore(dd[columns_to_check]))
outliers_zscore = z_scores > 3

outliers_iqr_count = outliers_iqr.sum()  # Count of outliers using IQR
outliers_zscore_count = outliers_zscore.sum()  # Count of outliers using Z-score

print("Outliers based on IQR (count per column):")
print(outliers_iqr_count)

print("\nOutliers based on Z-Score (count per column):")
print(outliers_zscore_count)
```

```python
# Clip outliers in the Volume column using the IQR method's lower and upper bounds
dd['Volume_clipped'] = dd['Volume'].clip(lower=lower_bound['Volume'], upper=upper_bound['Volume'])

# Check the result
print(dd[['Volume', 'Volume_clipped']].head())
```

```python
# Summary statistics of original and clipped Volume columns
print("Original Volume Statistics:")
print(dd['Volume'].describe())

print("\nClipped Volume Statistics:")
print(dd['Volume_clipped'].describe())
```

# Model Development


**Train-Test Split:** We preserved the temporal order while splitting the dataset into *80% training* and *20% testing*.

**Normalization:** *Min-Max Scaling* is applied to the features for consistent scaling.

**Model Selection:** Various regression models (L*inear Regression, Decision Tree, Random Forest, XGBoost*) are used.

**Model Training:** We train models and use *cross-validation* to ensure robustness.

**Hyperparameter Tuning:** We use *GridSearchCV* for tuning the *hyperparameters* of the *Random Forest model*.

**Model Evaluation:** The model performance is evaluated using *Mean Squared Error and R^2 Score*.

## Train-Test Split

```python
from sklearn.model_selection import train_test_split

X = dd[['Open', 'High', 'Low', 'Volume_clipped']]  
y = dd['Close'] 

train_size = int(len(dd) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

##  Normalize Numerical Features

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Min-Max Scaling (Scale features to [0, 1])
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Model Selection

### Linear Regression

```python
from sklearn.linear_model import LinearRegression

lin_reg_model = LinearRegression()
lin_reg_model.fit(X_train_scaled, y_train)
```

### Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor

# Initialize and train the model
dt_reg_model = DecisionTreeRegressor()
dt_reg_model.fit(X_train_scaled, y_train)
```

### Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor

# Initialize and train the model
rf_reg_model = RandomForestRegressor(n_estimators=100)
rf_reg_model.fit(X_train_scaled, y_train)
```

### Gradient Boosting Models (XGBoost)

```python
%pip install --upgrade xgboost scikit-learn
```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error, r2_score

xgb_model = xgb.XGBRegressor(objective='reg:squarederror', eval_metric='rmse')

# Train the model
xgb_model.fit(X_train_scaled, y_train)

# Make predictions
y_pred = xgb_model.predict(X_test_scaled)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")
# Mean Squared Error: 0.012782963125953914
# R^2 Score: 0.9482841001041672
```

## Model Training and Cross-Validation

```python
from sklearn.model_selection import cross_val_score

# Using Random Forest model for cross-validation example
cv_scores = cross_val_score(rf_reg_model, X_train_scaled, y_train, cv=5)  # 5-fold cross-validation

# Print the average cross-validation score
print(f"Random Forest CV Scores: {cv_scores}")
print(f"Mean CV Score: {cv_scores.mean()}")

# Random Forest CV Scores: [ 0.98535018  0.87192039  0.90381342  0.66453656 -2.18053665]
# Mean CV Score: 0.2490167824109597
```

## Hyperparameter Tuning (Grid Search)

```python
from sklearn.model_selection import GridSearchCV

# Define parameter grid for Random Forest
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

# Perform Grid Search with cross-validation
grid_search = GridSearchCV(estimator=RandomForestRegressor(), param_grid=param_grid, cv=5)
grid_search.fit(X_train_scaled, y_train)

# Best parameters and best score
print(f"Best Parameters: {grid_search.best_params_}")
print(f"Best Score: {grid_search.best_score_}")
```

##  Model Evaluation

```python
from sklearn.metrics import mean_squared_error, r2_score

# Predictions using Random Forest model
y_pred = rf_reg_model.predict(X_test_scaled)

# Evaluate using Mean Squared Error (MSE) and R^2 score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")
```

```python
import mlflow
import mlflow.sklearn
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Prepare features and target
X = dd[['Open', 'High', 'Low', 'Volume']]  # Features
y = dd['Close']  # Target

# Train-Test Split (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Start an MLflow experiment
with mlflow.start_run():
    # Train a RandomForestRegressor
    model = RandomForestRegressor(n_estimators=100)
    model.fit(X_train, y_train)
    
    # Predict
    y_pred = model.predict(X_test)
    
    # Calculate metrics
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = mse ** 0.5
    r2 = r2_score(y_test, y_pred)
    
    # Log metrics
    mlflow.log_metric("MAE", mae)
    mlflow.log_metric("RMSE", rmse)
    mlflow.log_metric("R2", r2)
    
    # Log the model
    mlflow.sklearn.log_model(model, "RandomForestModel")
    
    # Log parameters
    mlflow.log_param("Model Type", "RandomForestRegressor")
    mlflow.log_param("n_estimators", 100)
    
    print(f"Logged MAE: {mae}, RMSE: {rmse}, R2: {r2}")
```

```python
from mlflow.tracking import MlflowClient

# Register the best model in the MLflow Model Registry
client = MlflowClient()
model_uri = "runs:/<RUN_ID>/RandomForestModel"  # Replace <RUN_ID> with your actual run ID from the UI
model_name = "BestStockModel"

client.create_registered_model(model_name)  # Create the model registry entry
client.create_model_version(model_name, model_uri, "v1")

print(f"Model '{model_name}' registered successfully!")
```
