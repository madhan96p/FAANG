# Standarding

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

Columns = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']
dd[Columns] = scaler.fit_transform(dd[Columns])
```

# Dedecting outliers
**Purpose:** Outliers can affect the performance of machine learning models, especially regression models.

```python
# Dedecting outliers using Z-score
import numpy as np
from scipy import stats

# Z-score method to detect outliers
z_scores = stats.zscore(dd[Columns])
abs_z_scores = np.abs(z_scores)
dd_no_outliers = dd[(abs_z_scores < 3).all(axis=1)]  # Keep rows where Z-score < 3
```

## **Date/Time Features**

```python
dd['Date'] = pd.to_datetime(dd['Date'])

dd['Year'] = dd['Date'].dt.year
dd['Month'] = dd['Date'].dt.month
dd['Day'] = dd['Date'].dt.day
dd['Weekday'] = dd['Date'].dt.weekday  # 0 = Monday, 6 = Sunday
dd['Is_Weekend'] = dd['Weekday'].isin([5, 6])  # True if Saturday or Sunday
# Just checking for is there any wrong data
```

```python
dd['Is_Weekend'].value_counts()
```

```python
dd.drop(columns=['Is_Weekend'], inplace=True)
```

## Return Features
*predicting future stock prices.*

```python
dd.loc[dd['Date'] == '2005-01-03', 'Prev_Close'] = -0.857272 # First day has no previous close
dd['Return'] = (dd['Close'] - dd['Prev_Close']) / dd['Prev_Close'] * 100
# dd['Prev_Close'] = dd['Close'].shift(1)
# dd.loc[dd['Date'] == '2005-01-03', 'Return'] = 0  # First day has no return -0.857272
```

## Feature Relationships

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns
numeric_cols = dd.select_dtypes(include=[np.number])

# Correlation matrix
correlation_matrix = numeric_cols.corr()

plt.figure(figsize=(20, 16))  # Adjust as per your screen size
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()

```

```python
threshold = 0.7
high_corr = correlation_matrix[(correlation_matrix > threshold) | (correlation_matrix < -threshold)]
plt.figure(figsize=(16, 12))
sns.heatmap(high_corr, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Highly Correlated Features Heatmap")
plt.show()
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
from scipy.spatial.distance import squareform
import seaborn as sns
import matplotlib.pyplot as plt

# Convert correlation matrix to distance matrix
distance_matrix = 1 - correlation_matrix.abs()

# Fill NaN values with 0
distance_matrix = distance_matrix.fillna(0)

# Convert distance matrix to condensed form
condensed_dist_matrix = squareform(distance_matrix)

# Compute linkage matrix
linkage_matrix = linkage(condensed_dist_matrix, method='ward')

# Plot clustered heatmap
sns.clustermap(
    correlation_matrix,
    cmap="coolwarm",
    annot=True,
    fmt=".2f",
    figsize=(16, 12),
    row_linkage=linkage_matrix,
    col_linkage=linkage_matrix,
)

plt.title("Clustered Feature Correlation Heatmap")
plt.show()
```
