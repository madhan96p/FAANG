```python
# Standarding open, high, low, close, adj close, volume

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

Columns = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']
dd[Columns] = scaler.fit_transform(dd[Columns])
```

```python
display(df.describe())
display(dd.describe())
```

**Purpose:** Outliers can affect the performance of machine learning models, especially regression models.

```python
# Dedecting outliers using Z-score
import numpy as np
from scipy import stats

# Z-score method to detect outliers
z_scores = stats.zscore(dd[Columns])
abs_z_scores = np.abs(z_scores)
dd_no_outliers = dd[(abs_z_scores < 3).all(axis=1)]  # Keep rows where Z-score < 3
```

# Exploratory Data Analysis (EDA)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

plt.figure(figsize=(12, 6))

plt.plot(df['Date'], df['Open'], label='Open Price', color='blue')
plt.plot(df['Date'], df['Close'], label='Close Price', color='green')
plt.plot(df['Date'], df['High'], label='High Price', color='orange')
plt.plot(df['Date'], df['Low'], label='Low Price', color='red')

plt.title('Stock Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price ()')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
