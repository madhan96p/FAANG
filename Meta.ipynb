### *Remove Outliers:**
- **IQR Method:** We'll calculate the IQR for the Close price and Volume columns, and filter out values that are below the 25th percentile minus 1.5 times the IQR, or above the 75th percentile plus 1.5 times the IQR.

- **Z-score Method:** We'll compute the Z-scores for the Close price and Volume columns. Any value with a Z-score greater than 3 or less than -3 will be considered an outlier.

**IQR** and **Z-score** methods might not be ideal because stock prices can *fluctuate greatly* due to **market events**, and outliers in stock data are often justifiable as extreme market behavior.

```python
dr = dd
```

```python
import numpy as np
from scipy import stats

def remove_outliers_iqr(dr, column):
    Q1 = dr[column].quantile(0.25)
    Q3 = dr[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return dr[(dr[column] >= lower_bound) & (dr[column] <= upper_bound)]

def remove_outliers_zscore(dr, column):
    z_scores = np.abs(stats.zscore(dr[column]))
    return dr[z_scores < 3]

dr_iqr_cleaned = remove_outliers_iqr(dr, 'Close')
dr_iqr_cleaned = remove_outliers_iqr(dr_iqr_cleaned, 'Volume')

dr_zscore_cleaned = remove_outliers_zscore(dr, 'Close')
dr_zscore_cleaned = remove_outliers_zscore(dr_zscore_cleaned, 'Volume')

```

```python
print(f'Actual facebook dataset: {df.shape}')
print(f'Datas that needed: {dd.shape}')
print(f'Datas after cleaned by IQR: {dr_iqr_cleaned.shape}')
print(f'Datas after cleaned by Z-score: {dr_zscore_cleaned.shape}')
```
